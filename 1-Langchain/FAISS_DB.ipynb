```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import Chroma, FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings, OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter 


docs = TextLoader("speech.txt").load()
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=30)
final_docs = text_splitter.split_documents(docs)
print(final_docs)
```

```python
embeddings  = OllamaEmbeddings(model= "gemma:2b")
db = FAISS.from_documents(final_docs,embeddings)
db
```

```python
### query

query1 = "What does the speaker believe is the main reason the United States should enter the war?"
query2 = "How does the speaker describe the desired outcome of the war?"

docs1 = db.similarity_search(query1)
docs2 = db.similarity_search(query2)
```
