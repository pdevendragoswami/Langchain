# Character Text Splitter

```python
from langchain_community.document_loaders import TextLoader
docs = TextLoader("speech.txt").load()

from langchain_text_splitters import CharacterTextSplitter
text_splitter = CharacterTextSplitter(separator="\n\n", chunk_size=100,chunk_overlap=20)
final_text = text_splitter.split_documents(docs)
```

```python
print(final_text[0])
print(final_text[1])
```

```python
with open("speech.txt") as file:
    data = file.read()

text_splitter = CharacterTextSplitter(chunk_size=100, chunk_overlap = 20)
final_text = text_splitter.create_documents([data])
```

```python
print(final_text[0])
print(final_text[1])
```

As it was not able to find the separator, it create the chunk of larger size than the define size

HTML Header Text Splitter

```python
from langchain_text_splitters import HTMLHeaderTextSplitter

url = "https://plato.stanford.edu/entries/goedel/"
headers_to_split_on = [
    ("h1", "Header 1"),
    ("h2", "Header 3"),
    ("h3", "Header 3"),
    ("h4", "Header 4")

]
html_splitter  = HTMLHeaderTextSplitter(headers_to_split_on=headers_to_split_on)
html_header_split = html_splitter.split_text_from_url(url)
html_header_split
```

Json Splitter

```python
import json
import requests

json_data = requests.get("https://api.smith.langchain.com/openapi.json").json()

print(json_data)
```

```python
from langchain_text_splitters import RecursiveJsonSplitter

json_splitter = RecursiveJsonSplitter(max_chunk_size=300)

#loading into json format
json_chunks = json_splitter.split_json(json_data)
json_chunks
```

```python
for chunk in json_chunks[:3]:
    print(chunk)
```

```python
#loading into document format

json_chunks = json_splitter.create_documents([json_data])
json_chunks

for chunk in json_chunks[:3]:
    print(chunk)
```

```python
#loading into string/txt format

json_chunks = json_splitter.split_text(json_data)
for chunk in json_chunks[:3]:
    print(chunk)
```
