```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ["OPENAI_API_KEY"]= os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings
embedding = OpenAIEmbeddings(model="text-embedding-3-small")
embedding
```

```python
text = "this is a tutorial on OpenAI embedding"
```

```python
query_result = embedding.embed_query(text)
```

```python
print(query_result)
print(len(query_result))
```

```python
embedding_1024 = OpenAIEmbeddings(model="text-embedding-3-small",dimensions=1024)
query_result_1024 = embedding_1024.embed_query(text)
```

```python
print(query_result_1024)
print(len(query_result_1024))
```

```python
from langchain_community.document_loaders import TextLoader
docs = TextLoader("speech.txt").load()

from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_docs = text_splitter.split_documents(docs)
```

```python
final_docs
```

```python
#vector embedding and vector db
```

```python
from langchain_community.vectorstores import Chroma
db = Chroma.from_documents(final_docs,embedding_1024)
db
```

```python
#retrieve the result from query vector store db and
query = "It will be all the easier for us to conduct ourselves as belligerents"
retrieved_results = db.similarity_search(query)
print(retrieved_results)
```
